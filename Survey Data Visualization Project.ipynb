```python
import pandas as pd

# Load your CSV file
file_path = r"C:\Users\shiva\Downloads\sample_survey_data.csv"
df = pd.read_csv(file_path)

# Display the first few rows
print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='AgeGroup', data=df)
plt.title('Distribution of Age Groups')
plt.show()
```

```python
df['PreferredMode'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Preferred Mode of Learning')
plt.ylabel('')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
correlation = numeric_df.corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
from matplotlib.backends.backend_pdf import PdfPages

with PdfPages('survey_report.pdf') as pdf:
    # Bar chart
    sns.countplot(x='AgeGroup', data=df)
    plt.title('Distribution of Age Groups')
    pdf.savefig()
    plt.close()

    # Pie chart
    df['PreferredMode'].value_counts().plot.pie(autopct='%1.1f%%')
    plt.title('Preferred Mode of Learning')
    plt.ylabel('')
    pdf.savefig()
    plt.close()
```

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import streamlit as st

# Load CSV
df = pd.read_csv(r"C:\Users\shiva\Downloads\sample_survey_data.csv")

st.title(" Survey Data Visualization")

# Age Group Chart
st.subheader("Age Group Distribution")
fig1, ax1 = plt.subplots()
sns.countplot(x='AgeGroup', data=df, ax=ax1)
st.pyplot(fig1)

# Preferred Mode Pie Chart
st.subheader("Preferred Mode of Learning")
fig2, ax2 = plt.subplots()
df['PreferredMode'].value_counts().plot.pie(autopct='%1.1f%%', ax=ax2)
ax2.set_ylabel('')
st.pyplot(fig2)

# Heatmap
st.subheader("Correlation Heatmap")
fig3, ax3 = plt.subplots()
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm', ax=ax3)
st.pyplot(fig3)
```
